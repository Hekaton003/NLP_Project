```python
import json
import torch
import pandas as pd
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from sklearn.metrics import classification_report
```

```python
from huggingface_hub import login
login()
```

```python
model_name = "finki-ukim/VezilkaLLM"
```

```python
bitsandbytes_config = BitsAndBytesConfig(load_in_4bit=True,
                                         bnb_4bit_compute_dtype=torch.float16,
                                         bnb_4bit_quant_type='nf4',
                                         bnb_4bit_use_double_quant=True)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model= AutoModelForCausalLM.from_pretrained(model_name,
                                             quantization_config=bitsandbytes_config,
                                             device_map="auto",
                                             offload_folder="offload")
```

```python
def get_prompt(instruction,questions,contexts):
  prompts=[]
  for i in range(len(questions)):
    prompt = f"{instruction}\nContext:{contexts[i]}\nQuestion:{questions[i]}\nAnswer:"
    prompts.append(prompt)
  return prompts
```

```python
def training(prompts,num_tokens=1):
  predictions=[]
  for i in range(len(prompts)):
    input_tokens = tokenizer(prompts[i],return_tensors="pt").to("cuda:0")
    output_tokens = model.generate(**input_tokens,max_new_tokens=num_tokens,
                                   do_sample=False,
                                   pad_token_id=tokenizer.eos_token_id,
                                   eos_token_id=tokenizer.eos_token_id)

    response = tokenizer.decode(output_tokens[0], skip_special_tokens=True).strip()
    pred_answer = response.split("Answer:")[1]
    print(f"Iteration {i}\n")
    print(f"Response: {pred_answer}")
    predictions.append(pred_answer)

  return predictions
```

```python
ds = load_dataset(
    "json",
    data_files=[
        "hf://datasets/CounterBench/CounterBench/data_balanced_alpha_V1.json",
        "hf://datasets/CounterBench/CounterBench/data_balanced_backdoor_V2.json"
    ]
)

print(ds)
print(ds["train"].features)
```

```python
dataset = ds["train"].to_pandas()
```

```python
dataset
```

```python
dataset['type'].value_counts()
```

```python
instruction = f"Instruction: Answer the question strictly using the causal facts provided. Do not introduce any new information. For counterfactuals, simulate the scenario along the causal chain step by step. Only output the final answer as Yes or No. Do not write anything else."
```

## Basic Questions

```python
dataset_basic = dataset[dataset['type']=="basic"]
```

```python
dataset_basic
```

```python
questions = dataset_basic["question"].values.tolist()
contexts = dataset_basic["given_info"].values.tolist()
true_answers = dataset_basic["answer"].values.tolist()
```

```python
prompts = get_prompt(instruction,questions,contexts)
```

```python
prompts[0]
```

```python
predictions = training(prompts)
```

```python
predictions_new = [elem.lower().strip() for elem in predictions]
predictions_new
```

```python
print(classification_report(true_answers,predictions_new,zero_division=0))
```
